```
import tensorflow as tf
import glob
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image, ImageOps
from tensorflow import keras
from tensorflow.keras import layers
import time

```

```
#Generator model
def generator_model():
    model = keras.Sequential()
    model.add(layers.Dense(8*8*128, use_bias=False, input_shape=(100,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((8, 8, 128)))
    assert model.output_shape == (None, 8, 8, 128) # Note: None is the batch size

    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    assert model.output_shape == (None, 16, 16, 64)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(32, (5, 5), strides=(4, 4), padding='same', use_bias=False))
    assert model.output_shape == (None, 64, 64, 32)
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(3, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    assert model.output_shape == (None, 128, 128, 3)
    model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.RMSprop(0.0002,0.5))
    #model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(1e-4))

    return model
```

```
#test generator with niose
generator = generator_model()
noise = tf.random.normal([1, 100])
generated_image = generator.predict(noise)
generated_image=generated_image*127.5+127.5
plt.imshow(generated_image[0, :, :,  :], cmap='gray')
```

```
#discriminator model
def discriminator_model():
    model = tf.keras.Sequential()
    model.add(layers.Conv2D(32, (5, 5), strides=(2, 2), padding='same',
                                     input_shape=[128, 128, 3]))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.RMSprop(0.001,0.5))
    #model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(1e-4))
    

    return model
```

```
#Testing discriminator model with the output of generator
discriminator = discriminator_model()
decision = discriminator.predict(generated_image)
print (decision)
```

```
# GAN model combining generator and discriminator
def GAN_model(generator,discriminator):
  model = tf.keras.Sequential()
  model.add(generator)
  discriminator.trainable=False
  model.add(discriminator)
  model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.RMSprop(0.0002,0.5))
  #model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(1e-4))
  return model
```

```
#Testing GAN model with noise
gan=GAN_model(generator,discriminator)
noise = tf.random.normal([1, 100])
decision = gan.predict(noise)
print (decision)
```

```
#initializing test_input(noise) for outputs and directories
noise_dim = 100
num_examples_to_generate = 16
test_input = tf.random.normal([num_examples_to_generate, noise_dim])
data_root='/content/gdrive/My Drive/pokemon'
checkpoint_dir=data_root+"/training_checkpoints"
```

```
def train_model(data,n_epochs=EPOCHS,batch_size=BATCH_SIZE,weights=Continue_training):
 discriminator = discriminator_model()
 generator = generator_model()

 if weights!=0:
   #Loading previous trained models
		generator.load_weights(checkpoint_dir+'/g_mod_'+str(weights)+'.h5')
		discriminator.load_weights(checkpoint_dir+'/d_mod_'+str(weights)+'.h5')
 gan=GAN_model(generator,discriminator)
 n_batches=int(data.shape[0]/batch_size)
 #print(n_batches)

 for epoch in range(n_epochs):
   print("Epoch : {}/{}".format(epoch+1+weights, n_epochs+weights))

   for i in range(n_batches):
     #noise=np.random.rand(batch_size,100) #can use numpy or tensorflow
     noise=tf.random.normal([batch_size,100])
     #print(noise.shape)

     real_input=data[i*batch_size:(i+1)*batch_size]
     fake_input = generator.predict(noise)
     d_input=np.concatenate((real_input,fake_input))
     d_lables =np.concatenate((np.ones((batch_size,1)),np.zeros((batch_size,1))))
     discriminator.trainable = True
     d_loss = discriminator.train_on_batch(d_input,d_lables)

     discriminator.trainable = False
     g_loss = gan.train_on_batch(noise,np.ones((batch_size,1)))
     #g_loss = gan.train_on_batch(noise,np.ones((batch_size,1)))
     print("Generator Loss(Batch: {}): {}   discriminator Loss(Batch: {}): {}".format(i, g_loss, i, d_loss))
     #generate_and_save_images(generator, epoch)

   if (epoch+1)%10==0:
     #display output image for every 10 epoch
     generate_images(generator, epoch+weights)

   if (epoch+1)%100==0:
     #display and save output image for every 100 epoch
     generator.save_weights(checkpoint_dir+'/g_mod_'+str(epoch+1+weights)+'.h5',True)
     discriminator.save_weights(checkpoint_dir+'/d_mod_'+str(epoch+1+weights)+'.h5',True)
     generate_and_save_images(generator, epoch+weights)
```

```
def generate_images(model, epoch):
  pred = model.predict(test_input)
  #de-normalization
  pred=pred*127.5+127.5
  img=[]
  for i in range(4):
    img.append(cv2.hconcat([pred[(i*4)],pred[(i*4)+1],pred[(i*4)+2],pred[(i*4)+3]]))
  img1=cv2.vconcat([img[0],img[1],img[2],img[3]])
  #cv2.imshow('img',img1)
  #cv2_imshow is for Colob
  cv2_imshow(img1)
def generate_and_save_images(model, epoch):
  pred = model.predict(test_input)
  #de-normalization
  pred=pred*127.5+127.5
  img=[]
  #image concatination to form 4x4 grid
  for i in range(4):
    img.append(cv2.hconcat([pred[(i*4)],pred[(i*4)+1],pred[(i*4)+2],pred[(i*4)+3]]))
  img1=cv2.vconcat([img[0],img[1],img[2],img[3]])
  #cv2.imshow('img',img1)
  #cv2_imshow is for Colob
  cv2_imshow(img1)
  cv2.imwrite(data_root+"/output/img_"+str(epoch+1)+".jpg",img1)
```

```
EPOCHS = 6000
BATCH_SIZE = 273
data_root='/content/gdrive/My Drive/pokemon'
checkpoint_dir=data_root+"/training_checkpoints"

X_train = []
#Getting the list of images in the file
images = os.listdir(data_root+"/pokemon/pokemon_jpg")
#print(images)
for image in images:
  im=cv2.imread(data_root+"/pokemon/pokemon_jpg/"+image)
  im=cv2.resize(im,(128,128))
  im = np.asarray(im)
  im = (im.astype(np.float32)-127.5)/127.5
  X_train.append(im)
X_train = np.array(X_train)
#print(X_train.shape)
```

```
#enter last checkpoint number
Continue_training=0
train_model(X_train,n_epochs=EPOCHS,batch_size=BATCH_SIZE,weights=Continue_training)
```
